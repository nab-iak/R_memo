## 逐步回归

```r
mtcars
```

原理就是删除模型中的任意一个变量（或不删除）后得到模型的AIC，然后将模型AIC从小到大排序  
最上面就是拟合更好的模型  
并且基于第一个模型继续重复删除变量递归得到新模型的AIC，直到得到最小的AIC

```r
all <- lm(qsec~.,mtcars[,c(1,3:5,7)])
stats::step(all)
```

```r
{lm(qsec~mpg + disp + hp + drat,mtcars[,c(1,3:5,7)]) |> summary()}$adj.r.squared
{lm(qsec~mpg + hp + drat,mtcars[,c(1,3:5,7)]) |> summary()}$adj.r.squared
{lm(qsec~hp + drat,mtcars[,c(1,3:5,7)]) |> summary()}$adj.r.squared
```

# 用PRSA数据做回归预测
---

加载数据

```r
PRSA_path <- getwd() |> dirname() |> paste("/data/Beijing Multi-Site Air-Quality Data/data.csv",sep="") |> file.path()
PRSA_dt <- read.csv(PRSA_path)
head(PRSA_dt)
```

查看缺损值：没有

```r
summary(PRSA_dt)
```

Date没什么用，保留其他列  

```r
PRSA_dt <- PRSA_dt[,2:ncol(PRSA_dt)]
head(PRSA_dt)
```

检查各变量的相关性  
无语子，一片红

```r
cor(PRSA_dt,method="spearman")
cor(PRSA_dt,method="spearman") |> ggcorrplot::ggcorrplot()
```

```r
{cor(PRSA_dt,method="spearman") |> psych::corr.test()}
{cor(PRSA_dt,method="spearman") |> psych::corr.test()}$p
```

检查方差膨胀因子，看看共线性问题  
再多看一眼就会爆炸

```r
lm(Volume~.,PRSA_dt) |> summary()
lm(Volume~.,PRSA_dt) |> car::vif()
```

删掉vif最高的Close试试

```r
lm(Volume~.-Close,PRSA_dt) |> summary()
lm(Volume~.-Close,PRSA_dt) |> car::vif() 
```

先逐步回归

```r
lm(Volume ~ ., data = PRSA_dt) |> stats::step() |> summary()
```

再最优子集回归

```r
(models_summary <- leaps::regsubsets(Volume ~ High + Low + Adj.Close, data = PRSA_dt) |> summary())
str(models_summary)
```

```r
par(mfrow=c(2,2))
plot(models_summary$rss,type="line")
plot(models_summary$adjr2,type="line")
points(
    which.max(models_summary$adjr2),
    models_summary$adjr2[which.max(models_summary$adjr2)],
    color="black",pch=20,cex=2
    )
plot(models_summary$cp,type="line")
points(
    which.min(models_summary$cp),
    models_summary$cp[which.min(models_summary$cp)],
    color="black",pch=20,cex=2
    )
plot(models_summary$bic,type="line")
points(
    which.min(models_summary$bic),
    models_summary$bic[which.min(models_summary$bic)],
    color="black",pch=20,cex=2
    )
```

所以最佳的模型应该是 lm(Volume ~ High + Low + Adj.Close, data = PRSA_dt)

```r
lm(Volume ~ High + Low + Adj.Close, data = PRSA_dt) |> summary()
```

## 计算函数
[38.2.3.3 预测均方误差计算](https://www.math.pku.edu.cn/teachers/lidf/docs/Rbook/html/_Rbook/stat-learning.html) 给的例子  

```r
predict.regsubsets <- function(object, newdata, id, ...){
  form <- as.formula(object$call[[2]])
  mat <- model.matrix(form, newdata)
  coefi <- coef(object, id=id)
  xvars <- names(coefi)
  mat[, xvars] %*% coefi
}
```

### 各部分解释

object$call[[2]]表示的是object，也就是lm模型的方程式中的call的formula中`qsec~.`部分  
具体的可以看下面输出

```r
lm(qsec~.,mtcars[,c(1,3:5,7)])$call
lm(qsec~.,mtcars[,c(1,3:5,7)])$call[[1]]

lm(qsec~.,mtcars[,c(1,3:5,7)])$call[[2]]

lm(qsec~.,mtcars[,c(1,3:5,7)])$call[[3]]
```
